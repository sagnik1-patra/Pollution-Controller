```python
# 📦 Libraries
import numpy as np
import pandas as pd
import rasterio
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import random

# 📂 Load the TIFF
tif_path = r"C:\Users\sagni\Downloads\Pollution controller\Global_Mean_Chlorophyll_a_2019_2023.tif"

with rasterio.open(tif_path) as src:
    chlor_a = src.read(1)
    chlor_a = np.nan_to_num(chlor_a, nan=0)  # Replace NaNs with 0
    mean_chlor_a = np.mean(chlor_a)  # Global average chlor_a

# 📊 Create DataFrame of years and mean chlor_a
years = np.arange(2019, 2024)
data = pd.DataFrame({
    'Year': years,
    'Chlor_a': np.linspace(mean_chlor_a * 0.8, mean_chlor_a * 1.1, len(years))  # Simulated variation
})

# 🦠 Artificial Immune System (AIS) - Clone Selection (for noise and optimization)
def ais_optimize(data, clones=10, mutation_rate=0.05):
    best_fit = data.copy()
    for _ in range(clones):
        mutated = best_fit.copy()
        noise = np.random.normal(0, mutation_rate, size=len(mutated))
        mutated['Chlor_a'] += noise * mutated['Chlor_a']
        if mutated['Chlor_a'].std() < best_fit['Chlor_a'].std():
            best_fit = mutated
    return best_fit

# Apply AIS optimization
optimized_data = ais_optimize(data)

# 🌲 Random Forest for Trend Prediction
X = optimized_data['Year'].values.reshape(-1, 1)
y = optimized_data['Chlor_a'].values
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# 📈 Predict future years
future_years = np.arange(2024, 2035).reshape(-1, 1)
future_predictions = rf_model.predict(future_years)

# 📊 Combine for full trend
full_years = np.concatenate([years, future_years.flatten()])
full_values = np.concatenate([y, future_predictions])

# 📉 Plot trend and prediction
plt.figure(figsize=(10, 6))
plt.plot(years, y, 'bo-', label='Observed (AIS Optimized)')
plt.plot(future_years, future_predictions, 'ro--', label='Predicted (RF)')
plt.xlabel('Year')
plt.ylabel('Mean Chlorophyll-a')
plt.title('Global Chlorophyll-a Trend (2019–2034)')
plt.legend()
plt.grid(True)
plt.show()

# 📦 Save future predictions
future_df = pd.DataFrame({
    'Year': future_years.flatten(),
    'Predicted_Chlor_a': future_predictions
})
future_df.to_csv(r"C:\Users\sagni\Downloads\Pollution controller\Global_Future_Chlorophyll_a_2024_2034.csv", index=False)
print("✅ Future predictions saved to Global_Future_Chlorophyll_a_2024_2034.csv")
```
