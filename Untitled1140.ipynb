```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import optuna  # For AIS-like optimization
import pandas as pd

# 📂 Path to GeoTIFF
tif_path = r"C:\Users\sagni\Downloads\Pollution controller\Delhi_2023_AER_AI.tif"

# 🗺 Load GeoTIFF and compute mean AER_AI
with rasterio.open(tif_path) as src:
    data = src.read(1)
    masked_data = np.ma.masked_equal(data, src.nodata)  # Ignore nodata values
    mean_aer_ai_2023 = masked_data.mean()
    print(f"📌 Mean AER_AI for 2023: {mean_aer_ai_2023:.4f}")

# 📅 Simulate annual data (2014–2023) based on slight random variations around 2023 mean
np.random.seed(42)
years = np.arange(2014, 2024)
aer_ai_values = mean_aer_ai_2023 + np.random.normal(0, 0.05, size=len(years))

# 📈 Features and target
X = years.reshape(-1, 1)
y = aer_ai_values

# 🚀 AIS + Random Forest for trend prediction
def objective(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 500)
    max_depth = trial.suggest_int('max_depth', 3, 15)
    rf = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
    rf.fit(X, y)
    y_pred = rf.predict(X)
    return mean_squared_error(y, y_pred)

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30)

# ✅ Best params & final model
best_params = study.best_params
rf_final = RandomForestRegressor(**best_params, random_state=42)
rf_final.fit(X, y)

# 📅 Predict future (2024–2034)
future_years = np.arange(2024, 2035).reshape(-1, 1)
future_predictions = rf_final.predict(future_years)

# 📊 Plot original + predicted trend
plt.figure(figsize=(10, 6))
plt.plot(years, y, 'o-', label='Simulated Observed AER_AI (2014–2023)')
plt.plot(future_years, future_predictions, 's--', color='red', label='Predicted AER_AI (2024–2034)')
plt.xlabel('Year')
plt.ylabel('Mean AER_AI')
plt.title('Delhi AER_AI Trend & Future Prediction (AIS + Random Forest)')
plt.legend()
plt.grid(True)
plt.show()

# 📄 Save predictions
future_df = pd.DataFrame({'Year': future_years.flatten(), 'Predicted_AER_AI': future_predictions})
output_csv = r"C:\Users\sagni\Downloads\Pollution controller\Delhi_AER_AI_2024_2034.csv"
future_df.to_csv(output_csv, index=False)
print(f"✅ Future predictions saved to: {output_csv}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio
import pandas as pd

# 📂 Paths
tif_path = r"C:\Users\sagni\Downloads\Pollution controller\Delhi_2023_AER_AI.tif"
predicted_csv = r"C:\Users\sagni\Downloads\Pollution controller\Delhi_AER_AI_2024_2034.csv"

# 🗺 Load GeoTIFF
with rasterio.open(tif_path) as src:
    data = src.read(1)
    masked_data = np.ma.masked_equal(data, src.nodata)  # Mask nodata values
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

# 📖 Load predicted values
pred_df = pd.read_csv(predicted_csv)

# 🎨 Create figure
fig, ax = plt.subplots(figsize=(10, 8))

# 🔥 Plot base map (2023 data as background)
cmap = plt.cm.inferno
heatmap = ax.imshow(masked_data, extent=extent, cmap=cmap, origin='upper')
plt.colorbar(heatmap, ax=ax, label="AER_AI (2023)")

# 📝 Annotate future predictions
for i, row in pred_df.iterrows():
    ax.text(extent[0]+0.01, extent[3]-0.02-(i*0.01), f"{int(row['Year'])}: {row['Predicted_AER_AI']:.3f}", 
            color='white', fontsize=10, ha='left', va='top', bbox=dict(facecolor='black', alpha=0.5, pad=2))

# 📌 Titles and labels
ax.set_title("Delhi AER_AI Heatmap (2023 Base) + Future Predictions (2024–2034)", fontsize=14)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_xticks([])
ax.set_yticks([])

plt.show()
```
